##What if we have a set of resistances?

```
R_i = {'name':'R_i','type':'conv','h':9.2,'area':15}
```

```
ResistanceList = [R1,R2,R_i]
```

```
Rtot = 0
fancyResults = {'Rtot':0}
for anyR in ResistanceList:
  if anyR['type'] == 'cond':
    Rvalue = round(anyR['length']/anyR['conductivity']/anyR['area'],4)
  elif anyR['type'] == 'conv':
    Rvalue = round(1/anyR['h']/anyR['area'],4)
  anyR['Rvalue'] = Rvalue
  Rtot += Rvalue
fancyResults['Rtot'] = Rtot
```

```
fancyResults
```
