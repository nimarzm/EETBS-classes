```
def epsilonEffective(eps1, eps2):
  """This function is calculating the effective emissivity in an air gap"""
  result = 1/(1/eps1 + 1/eps2 - 1)
  return result
```

```
help(epsilonEffective)
```

```
result1 = epsilonEffective(0.1,0.1)
print(result1)
```

```
def epsilonEffective(eps1=0.9, eps2=0.9):
  """This function is calculating the effective emissivity in an air gap.
  By default, the emissivity of both surfaces are considered 0.9 in this function."""
  result = 1/(1/eps1 + 1/eps2 - 1)
  return result
```

```
epsilonEffective(0.1)
```

```
epsilonEffective(eps2=0.1)
```

```
help(epsilonEffective)
```

```
myEpsilons = [0.1,0.2,0.3,0.4,0.7,0.9]
```

```
for anyEpsilon2 in myEpsilons:
  result2 = epsilonEffective(eps2=anyEpsilon2)
  print(result2)
```

```
EpsEffectiveCalc = []
eps1 = 0.9
def epsilonEffectiveVectorizedOneSideNormalWall(ListOfEpsilons):
  for anyEps in ListOfEpsilons:
    result = 1/(1/eps1 + 1/anyEps - 1)
    EpsEffectiveCalc.append(result)
```

```
ListOfEpsilons = [0.2,0.4,0.6]
epsilonEffectiveVectorizedOneSideNormalWall(ListOfEpsilons)
EpsEffectiveCalc
```
