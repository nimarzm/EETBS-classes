##Problem: what if we have plenty of resistances?

**for loop:**

```
R1 = ["R_foam","cond",0.03,0.026,15]
R2 = ["R_plaster","cond",0.04,0.22,15]
R3 = ["R_wood","cond",0.05,0.8,15]
```

```
L1 = [1,3,7,9]
for mina in L1: #MINA IS A LOCAL VARIABLE
  print(mina)
```

```
ResistanceList = [R1,R2,R3]
ResistanceList
```

```
R_tot = 0
for anyR in ResistanceList:
  print(anyR[0])
  Rvalue_anyR = round(anyR[2]/anyR[3]/anyR[4],3)
  print(Rvalue_anyR)
  anyR.append(Rvalue_anyR)
  print(anyR)
  R_tot = R_tot + Rvalue_anyR
```

```
print(ResistanceList)
print()
print("The total resistance is " + str(R_tot) + " [degreeC/W]")
```

##Combining the convective and conductive in the same loop

```
R_in = ["R_internal", "conv", 10, 15]
R_out = ["R_external", "conv", 40, 15]
R1 = ["R_foam","cond",0.03,0.026,15]
R2 = ["R_plaster","cond",0.04,0.22,15]
R3 = ["R_wood","cond",0.05,0.8,15]
ResistanceList = [R_in,R1,R2,R3,R_out]
```

```
print(ResistanceList)
```

```
R_tot = 0
for anyR in ResistanceList:
  if anyR[1] == "cond":
    Rvalue_anyR = round(anyR[2]/anyR[3]/anyR[4],3)
    anyR.append(Rvalue_anyR)
  elif anyR[1] == "conv":
    Rvalue_anyR = round(1/anyR[2]/anyR[3],3)
    anyR.append(Rvalue_anyR)
  R_tot = R_tot + Rvalue_anyR   #It is more prof. to put the counter (R_tot) outside the loop.
```

```
print(ResistanceList)
print("The total resistance is " + str(round(R_tot,3)) + " [degreeC/W]")
```
