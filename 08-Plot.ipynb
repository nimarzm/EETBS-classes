#Week 8: Matplotlib



```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

##Basics

Creating a range of numbers:

```
X = np.arange(-10,10.1,0.1)
```

```
Y = X**2 - 5
```

```
plt.figure()
plt.plot(X,Y)
```

Another way of creating a range of numbers:

```
X2 = np.linspace(0,10,100)
```

```
fig = plt.figure() #by using this command, we put all the following plots, inside a figure (for further saving and etc.)
plt.plot(X2, np.sin(X2), "-")
plt.plot(X2, np.cos(X2), "--")
fig.savefig("MyPlot.png")
```

##MATLAB style sub-plots

```
plt.figure()

plt.subplot(3,1,1) #2 rows of plots, 1 column, index here is 1
plt.plot(X2, np.sin(X2),"--")

plt.subplot(3,1,2)
plt.plot(X2,np.cos(X2),"-.")

plt.subplot(3,1,3)
plt.plot(X2,np.sin(X2)*np.cos(X2),".")
```

More elaborative name: giving name to each axis:

```
fig,ax = plt.subplots(2)
```

```
ax[0]
```

```
ax[1]
```

By this mean, we can first biuld the entire plot, and then fill them with our desired plots:

```
ax[0].plot(X2, np.sin(X2), "-")
ax[1].plot(X2, np.cos(X2), "-")
fig
```

##Colors/line style

Colors:

```
plt.figure()
plt.plot(X2, np.sin(X2), color="blue")
plt.plot(X2, np.sin(X2-1), color="g")
plt.plot(X2, np.sin(X2-2), color="y")
plt.plot(X2, np.sin(X2-3), color="0.75") #using gray scale btwn 0 and 1
plt.plot(X2, np.sin(X2-4), color="#FF0135")
plt.plot(X2, np.sin(X2-5), color=(1.0,0.2,0.3)) #RGB
plt.plot(X2, np.sin(X2-6), color="chartreuse") #html color codes
```

Line style:

```
plt.figure()

plt.plot(X2,X2,linestyle="solid")
plt.plot(X2,X2+1,linestyle="dashed")
plt.plot(X2,X2+2,linestyle="dashdot")
plt.plot(X2,X2+3,linestyle="dotted")
```

You can use the shortcuts:

```
plt.figure()

plt.plot(X2,X2,linestyle="-")
plt.plot(X2,X2+1,linestyle="--")
plt.plot(X2,X2+2,linestyle="-.")
plt.plot(X2,X2+3,linestyle=":")
```

Combine style and color:

```
plt.figure()

plt.plot(X2,X2,"-g")
plt.plot(X2,X2+1,"--r")
plt.plot(X2,X2+2,"-.y")
plt.plot(X2,X2+3,":b")
```

##Axis/title/legend

Configuring the axis ranges:

```
fig4 = plt.figure()

plt.plot(X2,X2,"-g")
plt.plot(X2,X2+1,"--r")
plt.plot(X2,X2+2,"-.y")
plt.plot(X2,X2+3,":b")
plt.xlim(2,5)
plt.ylim(2,6)
```

Adding x and y labels:

```
fig5 = plt.figure()

plt.plot(X2,X2,"-g")
plt.plot(X2,X2+1,"--r")
plt.plot(X2,X2+2,"-.y")
plt.plot(X2,X2+3,":b")
plt.xlim(2,5)
plt.ylim(2,6)
plt.xlabel("This is X")
plt.ylabel("This is Y")
```

Adding title for the chart:

```
fig6 = plt.figure()

plt.plot(X2,X2,"-g")
plt.plot(X2,X2+1,"--r")
plt.plot(X2,X2+2,"-.y")
plt.plot(X2,X2+3,":b")
plt.xlim(2,5)
plt.ylim(2,6)
plt.xlabel("This is X")
plt.ylabel("This is Y")
plt.title("This is the title")
```

Add lablels to multiple graphs in the same plot (legend):

```
fig7 = plt.figure()

plt.plot(X2,np.sin(X2),"-g",label="sin(x)")
plt.plot(X2,np.cos(X2),"-.r",label="cos(x)")
plt.legend()
```

##Scatter plot

Still using the plot (plt.plot()) function:

```
fig8 = plt.figure()
plt.plot(X2,np.sin(X2),"d",color="red")
```

Options for markers: o, x, >, <, * and etc.:

```
fig9 = plt.figure()
plt.plot(X2,np.sin(X2),"*",color="red")
```

You can have both the marker and the plot:

```
fig10 = plt.figure()
plt.plot(X2,np.sin(X2),"-ob") #"-" is the style, "o" is the scattered, "b" is black color.
```

Using scatter command:

```
plt.figure()
plt.scatter(X2,np.sin(X2),marker="o")
```

##Check all the properties

```
fig11 = plt.figure()
plt.plot(X2,np.sin(X2),"o-",color="black",markersize=5,linewidth=1,markerfacecolor="red",markeredgecolor="gray",markeredgewidth=2)
```

##Create fancy plot

We will first create some random numbers:

```
rng=np.random.RandomState(10)
x=rng.randn(100)
y=rng.randn(100)
colors=rng.randn(100)
sizes=1000*rng.randn(100)
plt.scatter(x,y,c=colors,s=sizes,alpha=0.3,cmap="viridis")
plt.colorbar()
```

##Using plots for our application

Bar chart:

```
item1 = [1,3,5]
item2 = [2,4,6]
labels = ["wall","floor","door"]
heatingLoad = [2581,3500,300]
coolingLoad = [2300,3400,250]
```

```
fig1 = plt.figure()
plt.bar(item1,heatingLoad,label="heating load",color="red")
plt.bar(item2,coolingLoad,label="cooling load",color="blue")
plt.legend()
plt.xticks(item1,labels)
```

Pie chart:

```
slices=[7,2,2,13]
activities = ["sleeping","eating","working","hobbies"]
cols=["c","m","r","b"]
plt.pie(slices,labels=activities, colors=cols,startangle=90,shadow=True,explode=(0,0.1,0,0),autopct="%1.1f%%")
```
