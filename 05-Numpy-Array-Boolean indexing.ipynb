##How do we use it for our applications?

```
opaque_item_list = ["wall","ceiling","door"]
opaque_item_array = np.array(opaque_item_list)
opaque_U_list = [0.438,0.25,1.78]
opaque_U_array = np.array(opaque_U_list)
```

```
opaque_area_array = np.array([105.8, 200, 2.2])
```

```
T_inside_building = 20
T_outside = -4.8
deltaT_heating = T_inside_building - T_outside
```

```
opaque_HF_array = deltaT_heating*opaque_U_array
opaque_HF_array
```

```
opaque_Q_array = opaque_HF_array * opaque_area_array
opaque_Q_array
```

```
opaque_Q_array_rounded = np.round(opaque_Q_array,2)
opaque_Q_array_rounded
```

Extracting data from arrays.

```
Q_ceiling = opaque_Q_array_rounded[1]
Q_ceiling
```

```
Q_door = opaque_Q_array_rounded[-1]
Q_door
```

##Using Numpy for our resistance problem

```
resistance_name = np.array(["R_internal", "R_foam", "R_wood", "R_plaster", "R_external"])
resistance_type = np.array(["conv", "cond", "cond", "cond", "conv"])
resistance_A = np.array([15,15,15,15,15])
resistance_k = np.array([None,0.05,0.4,1,None])
resistance_L = np.array([None, 0.06, 0.1, 0.01, None])
resistance_h = np.array([8.9, None, None, None, 20])
```

```
resistance_Rvalue = np.zeros(5)
resistance_Rvalue
```

To calculate the Rvalues:
First we create an array of booleans, that tells us which resistances are conductive

```
bool_resistance_cond = resistance_type == "cond"
bool_resistance_cond
```

```
resistance_L_cond = resistance_L[bool_resistance_cond]
resistance_L_cond
```

```
resistance_k_cond = resistance_k[bool_resistance_cond]
resistance_k_cond
```

```
resistance_A_cond = resistance_A[bool_resistance_cond]
resistance_A_cond
```

```
resistance_R_cond = resistance_L_cond/resistance_k_cond/resistance_A_cond
resistance_R_cond
```

```
resistance_Rvalue[bool_resistance_cond] = resistance_R_cond
```

```
resistance_Rvalue
```

```
bool_resistance_conv = resistance_type == "conv"
bool_resistance_conv
```

```
resistance_A_conv = resistance_A[bool_resistance_conv]
resistance_A_conv
```

```
resistance_h_conv = resistance_h[bool_resistance_conv]
resistance_h_conv
```

```
resistance_Rvalue_conv = 1.0/resistance_h_conv/resistance_A_conv
resistance_Rvalue_conv
```

```
resistance_Rvalue[bool_resistance_conv] = resistance_Rvalue_conv
resistance_Rvalue
```
