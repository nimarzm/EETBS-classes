#Conditionals

```
material = "glass"
if material == "glass": #remember to use "==" instead of "="
    k = 0.78 #W/mK
else:
  print("!!")

print(k)

```

```
material = "glass"
if material == "glass": #remember to use "==" instead of "="
  k = 0.78 #W/mK
elif material == "brick":
  k = 0.4 #W/mK
else:
  print("!!")

print(k)
```

#Nested **if**


```
material = "glass"
type_glass = 2
if material == "glass": #remember to use "==" instead of "="
  if type_glass == 1:
    k = 0.78 #W/mK
  elif type_glass == 2:
    k = 0.89 #W/mK
  else:
    print("!")
elif material == "brick":
  k = 0.4 #W/mK
else:
  print("!!")

print(k)
```
